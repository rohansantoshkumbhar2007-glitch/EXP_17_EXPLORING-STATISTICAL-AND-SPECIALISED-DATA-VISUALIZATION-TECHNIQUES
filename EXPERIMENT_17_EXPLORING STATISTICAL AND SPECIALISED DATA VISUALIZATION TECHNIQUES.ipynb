```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import numpy as np

np.random.seed(0)
df = pd.DataFrame({
    'Category':['A','B','C','D','E'],
    'Value':[50,70,30,90,60],
    'Sales': np.random.randint(100,500,5),
    'Profit': np.random.randint(1000,5000,5)
})
print(df.head())
```

```python
#1. Area Plot
plt.figure()
plt.title('Area Plot')
plt.fill_between(df['Category'],df['Value'],color='purple',alpha=0.5)
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
#2.Pie Chart
plt.figure()
plt.title('Pie Chart')
plt.pie(df['Value'],labels=df['Category'],autopct='%2.2f%%')
plt.show()
```

```python
#Donut Chart
plt.figure()
plt.title('Donut Chart')
plt.pie(df['Value'],labels=df['Category'],autopct='%1.1f%%')
plt.pie([1],colors='w',radius=0.5)
fig=plt.gcf()
fig.gca().add_artist(plt.Circle((0,0),radius=0.5,fc='black'))
plt.show()
```
